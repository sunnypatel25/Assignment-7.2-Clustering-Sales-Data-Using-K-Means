```python
# Import the pandas library for data manipulation and analysis
import pandas as pd
# URL of the CSV database file containing weekly sales transaction data
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Applied-Artificial-Intelligence-Workshop/master/Datasets/Sales_Transactions_Dataset_Weekly.csv'
# Read the CSV file from the URL into a pandas DataFrame
df = pd.read_csv(file_url)
# Display the original DataFrame
df
```

```python
# Create a new DataFrame by dropping the first 55 columns
# These columns are removed to keep only relevant numeric features for clustering
df2 = df.drop(df.iloc[:, 0:55], inplace = False, axis = 1)
# Display the DataFrame used for clustering
df2
```

```python
# Import KMeans clustering algorithm from scikit-learn
from sklearn.cluster import KMeans
# Initialize the KMeans model with 8 clusters and a fixed 8 random state for reproducibility
k_means_model = KMeans(n_clusters=8, random_state=8)
# Fit the KMeans model on the prepared dataset
k_means_model.fit(df2)
# Extract the cluster labels assigned to each data point
labels = k_means_model.labels_
# Display the cluster labels
labels
```

```python
# Drop all columns from index 53 onward from the original DataFrame
# This removes weekly sales columns that are no longer needed
df.drop(df.iloc[:, 53:], inplace = True, axis = 1)
# Remove the Product_Code column since it is not needed for aggregation
df.drop('Product_Code', inplace = True, axis = 1)
# Add the cluster labels as a new column in the DataFrame
df['label'] = labels
# Display the DataFrame with assigned cluster labels
df
```

```python
# Aggregate the sales data by cluster label and calculate the sum for each cluster
df_agg = df.groupby('label').sum()
# Count the number of products in each cluster using column W0
df_final = df[['label','W0']].groupby('label').count()
# Rename the column to indicate it represents the number of products
df_final=df_final.rename(columns = {'W0':'count_product'})
# Calculate total sales per cluster by summing all weekly sales
df_final['total_sales'] = df_agg.sum(axis = 1)
# Calculate average yearly sales per product in each cluster
df_final['yearly_average_sales']= df_final['total_sales'] / df_final['count_product']
# Sort clusters by average yearly sales in descending order
df_final.sort_values(by='yearly_average_sales', ascending=False, inplace = True)
# Display the final summarized DataFrame
df_final
```
